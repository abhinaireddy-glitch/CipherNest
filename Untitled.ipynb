```python
import pandas as pd
from sklearn.model_selection import train_test_split

file_path = r"C:\Users\HP\Downloads\MachineLearningCVE\Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv"

df = pd.read_csv(file_path)

# Remove leading/trailing spaces from all column names
df.columns = df.columns.str.strip()

print("Columns after cleaning:", df.columns.tolist())

# Stratified sampling
sampled_df, _ = train_test_split(
    df,
    train_size=30000,
    stratify=df['Label'],
    random_state=42
)

# Save result
sampled_df.to_csv("sampled_30k_dataset.csv", index=False)

print("Sampled dataset size:", len(sampled_df))
print(sampled_df['Label'].value_counts())
```
